```python
data = get_data_author(full_df)
categories = list(data.columns)[1:]
colors = resize_list(Category20_20, len(categories))

source = ColumnDataSource(data=data)

p = figure(
    plot_height=350,
    title="Song Counts by Year",
    toolbar_location=None, tools="")

renderers = p.vbar_stack(
    categories, x='year', width=0.9, color=colors, source=source,
    legend=[value(x) for x in categories],
    name=categories)

for r in renderers:
    cat = r.name
    hover = HoverTool(tooltips=[
        ("year", '@year'),
        ("category", cat),
    ], renderers=[r])
    p.add_tools(hover)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
