### Plotting

```python
from bokeh.core.properties import value
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.palettes import Category20_20
from bokeh.models import HoverTool
```

```python
categories = ['one', 'two'] * 20
```

```python
[value(x) for x in categories if len(categories) < 10]
```

```python
def make_bar_chart(albums_df, songs_df, cross='author'):
    data = get_album_data(albums_df, songs_df, cross)
    categories = list(data.columns)[1:]
    colors = resize_list(Category20_20, len(categories))

    source = ColumnDataSource(data=data)

    p = figure(
        plot_height=350,
        title="Song Counts by Year",
        toolbar_location=None, tools="")
    
    leg = [value(x) for x in categories] if len(categories) < 10 else None
    
    renderers = p.vbar_stack(
        categories, x='year', width=0.9, color=colors, source=source,
        legend=leg,
        name=categories)

    for r in renderers:
        cat = r.name
        hover = HoverTool(
            tooltips=[("year", '@year'), (cross, cat),
                      ("songs", "@{%s}" % cat)],
            renderers=[r])
        p.add_tools(hover)

    p.y_range.start = 0
    p.x_range.range_padding = 0.1
    p.xgrid.grid_line_color = None
    p.outline_line_color = None
    p.legend.location = "top_left"
    p.legend.orientation = "horizontal"

    show(p)
```

```python
make_bar_chart(albums_df, songs_df, cross='album')
```
