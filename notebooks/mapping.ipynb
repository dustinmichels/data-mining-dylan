```python
import json
import pandas as pd
import numpy as np
import folium
```

```python
with open('info.geojson') as f:
    data = json.load(f)
    
geojson = json.dumps(data)
```

```python
city_df = pd.read_json('../data/city_counts.json')
```

## Folium

```python
m = folium.Map(location=[20,-40], tiles="Mapbox Bright", zoom_start=2)

popup_text = ("<b>{}</b>, {} mention(s)<br>"
              "<i>Appears in:</i> {}")

for idx, row in city_df.iterrows():
    
    songs = ', '.join(row.songs)
    
    folium.CircleMarker(
        location=[row.lat, row.lng],
        radius=row['cnt']*1.5,
        fill=True,
        popup=folium.Popup(
            html=popup_text.format(row.city, row.cnt, songs),
            max_width=250),
    ).add_to(m)

m
```

```python
m.save('map.html')
```
